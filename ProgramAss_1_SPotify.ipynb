*    What are the mean, median, and standard deviation of the streams column?




```python
# Convert the 'streams' column to a numeric data type
# 'errors="coerce"' will force any non-convertible values (like strings) to be set as NaN (Not a Number)
spotify_2023['streams'] = pd.to_numeric(spotify_2023['streams'], errors='coerce')

# Calculate the mean (average) of the 'streams' column
mean_streams = spotify_2023['streams'].mean()

# Calculate the median (middle value) of the 'streams' column
median_streams = spotify_2023['streams'].median()

# Calculate the standard deviation of the 'streams' column (how spread out the numbers are)
std_streams = spotify_2023['streams'].std()

# Print the mean, median, and standard deviation values of the 'streams' column
print(f"Mean: {mean_streams}, Median: {median_streams}, Std Dev: {std_streams}")
```

# 4. **Temporal Trends**

*    Analyze the trends in the number of tracks released over time. Plot the number of tracks released per year







```python
# Group the data by 'released_year' and count the number of tracks ('track_name') for each year
# This creates a trend of how many tracks were released per year
trending_year = spotify_2023.groupby('released_year')['track_name'].count()

# Plot the trend of tracks released per year using a line plot
# 'kind="line"' specifies a line plot, and various styling options are used for customization
trending_year.plot(kind='line', color='g', marker='o', linestyle='-', linewidth=2)

# Add a title and axis labels to the plot for better understanding
plt.title('Number of Tracks Released per Year', fontsize=16)  # Plot title
plt.xlabel('Year', fontsize=12)  # X-axis label
plt.ylabel('Number of Tracks', fontsize=12)  # Y-axis label

# Add gridlines to the plot for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adjust the layout to ensure the plot elements are well-fitted
plt.tight_layout()

# Display the plot
plt.show()
```

The number of tracks released per year remained **relatively stable until the early 2000s**, after it began to increase gradually. From 2010, the growth increased significantly, **peaking around 2020.**



*  Does the number of tracks released per month follow any noticeable patterns? Which
month sees the most releases?





```python
# Group the data by 'released_month' and count the number of tracks ('track_name') for each month
most_release = spotify_2023.groupby('released_month')['track_name'].count()

# Plot the trend of tracks released per month using a line plot
most_release.plot(kind='line', color='g', marker='o', linestyle='-', linewidth=2)

# Add a title and axis labels to the plot for better understanding
plt.title('Number of Tracks Released per Month', fontsize=16)  # Plot title
plt.xlabel('Month', fontsize=12)  # X-axis label
plt.ylabel('Number of Tracks', fontsize=12)  # Y-axis label

# Add gridlines to the plot for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Customize the x-ticks to display month names instead of numbers
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                         'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Identify the month with the maximum number of releases
max_month = most_release.idxmax()  # Get the month with the most releases

# Draw a vertical dashed line to indicate the month with the most releases
plt.axvline(x=max_month, color='red', linestyle='--', linewidth=1.5, label=f'Most Releases ({max_month})')

# Add a legend to the plot
plt.legend()

# Adjust the layout to ensure the plot elements are well-fitted
plt.tight_layout()

# Display the plot
plt.show()
```

While there are **no consistent patterns in the number of tracks released each month**, the distribution indicates that the **highest number of track releases occurs in January, followed closely by May.**

# 5. **Genre and Music Characteristics**

```python
# Select relevant columns for correlation analysis
# These columns include 'streams', 'bpm', 'danceability_%', and 'energy_%'
columns_of_interest = ['streams', 'bpm', 'danceability_%', 'energy_%']

# Calculate the correlation matrix for the selected columns
# This matrix shows the pairwise correlation coefficients between the variables
correlation_matrix = spotify_2023[columns_of_interest].corr()

# Plot the correlation matrix using a heatmap for better visualization
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
# 'annot=True' adds the correlation coefficient values in the cells
# 'cmap' specifies the color palette, while 'vmin' and 'vmax' set the range of values

# Add a title to the heatmap
plt.title('Correlation Matrix: Streams vs Musical Attributes')

# Display the heatmap
plt.show()
```

None of the musical attributes, including **bpm**, **danceability_%**, and **energy_%**, exhibit strong correlations with the number of streams. The highest correlation observed is a modest **-0.105** between **danceability_%** and **streams**, indicating a slight inverse relationship.

```python
# Define the subplots with a 1x2 layout (1 row and 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 4))  # Set the size of the figure

# Analyze the correlation between 'danceability_%' and 'energy_%'
columns_of_interest = ['danceability_%', 'energy_%']  # Select the relevant columns
correlation_matrix = spotify_2023[columns_of_interest].corr()  # Calculate the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, ax=axs[0])
# Create a heatmap for the first set of variables with annotations and a color map
axs[0].set_title('Correlation: Danceability_% vs Energy_%')  # Set the title for the first subplot

# Analyze the correlation between 'valence_%' and 'acousticness_%'
columns_of_interest = ['valence_%', 'acousticness_%']  # Select the relevant columns
correlation_matrix = spotify_2023[columns_of_interest].corr()  # Calculate the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, ax=axs[1])
# Create a heatmap for the second set of variables with annotations and a color map
axs[1].set_title('Correlation: Valence_% vs Acousticness_%')  # Set the title for the second subplot

# Adjust the layout to ensure the subplots fit well within the figure
plt.tight_layout()

# Display the subplots
plt.show()

```

Danceability_% and Energy_% have a **weak positive correlation** ***(0.198)***.

Valence_% and Acousticness_% have a **weak negative correlation** ***(-0.082)***.

# 7. **Advanced Analysis**

```python
# Remove rows where the 'key' column has NaN values
# This ensures that only tracks with a defined musical key are considered
nan_key = spotify_2023.dropna(subset='key')

# Set a multi-index using 'track_name', 'key', and 'mode', and sort the DataFrame by 'streams' in descending order
getting_tracks = nan_key.set_index(['track_name', 'key', 'mode']).sort_values(by='streams', ascending=False)

# Display the DataFrame with tracks indexed by name, key, and mode
getting_tracks

# Group the data by 'key' and 'mode', summing the total streams for each group
grouped_data = nan_key.groupby(['key', 'mode'])['streams'].sum().reset_index()

# Create a bar plot to visualize total streams by key and mode
plt.figure(figsize=(14, 8))  # Set the figure size for better visibility
sns.barplot(data=grouped_data, x='key', y='streams', hue='mode')
# 'hue' distinguishes between different modes (major/minor) within the same key

# Add title and axis labels for clarity
plt.title('Total Streams by Key and Mode', fontsize=16)  # Title of the plot
plt.xlabel('Musical Key', fontsize=14)  # X-axis label
plt.ylabel('Total Streams', fontsize=14)  # Y-axis label

# Display the legend to indicate the mode represented by different colors
plt.legend(title='Mode', loc='upper right', fontsize=12)

# Adjust the layout to ensure all plot elements fit well
plt.tight_layout()

# Display the plot
plt.show()
```

Based on the streams data, tracks within **C# regardless of its mode gained the most streams**. Moreover, in general, **major keys gained much more streams in comparison with minor keys**

```python
# Define columns for playlists and charts to analyze their appearances
playlist_cols = ['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']
chart_cols = ['in_spotify_charts', 'in_apple_charts', 'in_deezer_charts', 'in_shazam_charts']

# Convert relevant playlist and chart columns to numeric values, coercing non-numeric entries to NaN
spotify_2023[playlist_cols + chart_cols] = spotify_2023[playlist_cols + chart_cols].apply(pd.to_numeric, errors='coerce')

# Calculate total appearances in playlists by summing across the specified columns
spotify_2023['total_playlist_appearances'] = spotify_2023[playlist_cols].sum(axis=1)

# Calculate total appearances in charts by summing across the specified columns
spotify_2023['total_chart_appearances'] = spotify_2023[chart_cols].sum(axis=1)

# Group the DataFrame by artist and sum their total playlist and chart appearances
artist_appearances = spotify_2023.groupby('artist(s)_name', as_index=False)[['total_playlist_appearances', 'total_chart_appearances']].sum()

# Calculate the overall total appearances by adding playlist and chart appearances
artist_appearances['total_appearances'] = artist_appearances['total_playlist_appearances'] + artist_appearances['total_chart_appearances']

# Get the top 10 artists based on total appearances
top_artists = artist_appearances.nlargest(10, 'total_appearances')

# Reshape the data for easier plotting using the melt function
top_artists_melted = top_artists.melt(id_vars='artist(s)_name',
                                      value_vars=['total_appearances', 'total_playlist_appearances', 'total_chart_appearances'],
                                      var_name='Appearance Type',
                                      value_name='Count')

# Create a bar plot to visualize the top 10 artists by playlist and chart appearances
plt.figure(figsize=(14, 8))  # Set the figure size
sns.barplot(data=top_artists_melted, x='artist(s)_name', y='Count', hue='Appearance Type')
# 'hue' differentiates between total appearances, playlist appearances, and chart appearances

# Add titles and labels for clarity
plt.title('Top 10 Artists by Playlist and Chart Appearances', fontsize=16)
plt.xlabel('Artist Name', fontsize=14)  # X-axis label
plt.ylabel('Appearance Count', fontsize=14)  # Y-axis label

# Display the legend to identify the different appearance types
plt.legend(title='Appearance Type', loc='upper right', fontsize=12)

# Rotate x-ticks for better readability
plt.xticks(rotation=45)

# Adjust the layout to ensure all elements fit well within the figure
plt.tight_layout()

# Display the plot
plt.show()
```

The data indicates that **artists consistently appear in playlists more than on charts**. Moreover, the 10 most frequently appearing artist are **The Weeknd, Taylor Swift, Ed Sheeran, Harry Styles, Eminem, Arctic Monkeys, Coldplay, Avicii, Adele, and both Dr. Dre and Snoop Dogg** on the 10th spot.On the other hand, Taylor Swift leads the artist appearances on charts even though The Weeknd has the most playlist and chart appearances in total.
